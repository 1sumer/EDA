### Step 10: Data Transformation
Before we proceed to Bi-variate Analysis, Univariate analysis demonstrated the data pattern as some variables to be transformed.

Price and Kilometer-Driven variables are highly skewed and on a larger scale. Let’s do log transformation.

Log transformation can help in normalization, so this variable can maintain standard scale with other variables:

- The primary goal is to make the data more suitable for analysis and modeling, as many statistical methods assume a normal distribution of the data.
- Skewed data, where the majority of the values are concentrated at one end, can lead to biased analyses.
- Log transformation helps in reducing the impact of extreme values, making the distribution more symmetric


**Logarithmic transformation** is a technique used to transform data that is not normally distributed into a normal distribution.
1. Log Transformation: Transform the response variable from y to log(y).

2. Square Root Transformation: Transform the response variable from y to √y.

3. Cube Root Transformation: Transform the response variable from y to y1/3.


Python, you can perform logarithmic transformation using the numpy.log() function 

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()

log_transform(data,['Kilometers_Driven','Price'])
```
