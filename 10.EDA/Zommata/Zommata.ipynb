```python
plt.figure(figsize=(10,20))
for i in range(len(columns_lst)):
    plt.subplot(5,3,i+1)
    plt.title(columns_lst[i])
    plt.xticks(rotation=45)
    plt.hist(df[columns_lst[i]])
    
plt.tight_layout()
```

```python
RATING = df['rate'].value_counts()
RATING
```

```python
RATING.plot(kind='pie',radius=2)
plt.title('rating')
```

```python
PLT = df[df['rate']==3.9].value_counts()[:20]
sns.set_style(style='whitegrid')
plt.figure(figsize=(5,5))
sns.distplot(PLT, label='Rate', color='tomato', kde=True)
plt.title('RATING')
plt.legend(loc=1)
plt.grid()
```
